```python
#3)Să se determine produsul scalar a doi vectori rari care conțin numere reale.
#Un vector este rar atunci când conține multe elemente nule. Vectorii pot avea oricâte
#dimensiuni. De ex. produsul scalar a 2 vectori unidimensionali [1,0,2,0,3] și [1,2,0,3,1]
#este 4.

#O(k)
#vf inainte de inmult daca elem sunt nenule, k nr elem
'''
input: x si y vectori
output: produs scalar(int)
'''
def Vect1(x, y):
    return sum(x[i] * y[i] for i in range(len(x)) if x[i] != 0 and y[i] != 0)

#O(n)
'''
input: x si y vectori
output: produs scalar(int)
'''
def Vect2(x,y):
 return sum(a * b for a, b in zip(x, y))

#testare

def Test3(cerinta):
    print("Testare")
    assert cerinta([1, 0, 2, 0, 3], [1, 2, 0, 3, 1]) == 4
    assert cerinta([0, 0, 0, 0, 0], [1, 2, 0, 3, 1]) == 0
    assert cerinta([1, 2], [3, 4, 5]) == 11
    assert cerinta([1, 0, 0, 4], [0, 2, 3, 1]) == 4
    print("Succes!\n")

#parcurgem si adunam prod elem coresp
#Copilot
#O(n)

def Copilot(l1,l2):
  rez = 0
  for i in range(len(l1)):
    rez += l1[i]*l2[i]
  return rez

Test3(Vect1)
Test3(Vect2)
Test3(Copilot)

x = [1, 0, 2, 1, 3]
y = [1, 2, 0, 3, 1]

print("Metoda 1: ",Vect1(x,y))
print("Metoda 2: ",Vect2(x,y))
print("Copilot: ", Copilot(x,y))
```

```python
from collections import defaultdict

#4)Să se determine cuvintele unui text care apar exact o singură dată în acel text. De ex.
#cuvintele care apar o singură dată în ”ana are ana are mere rosii ana" sunt: 'mere' și
#'rosii'.

'''
Vector de frecventa
O(n)
n numarul de cuvinte
se va face split pe cuvinte
vom parcurge fiecare cuvant din prop si verificam daca cuvantul a mai aparut in vector si
incrementam valoarea frecventei altfel adaugam noul cuvant si ii dam frecventa 1
stocam cuvintele cu frecventa 1 in lista finala

input: sir de caractere(string)
output: lista cuvintelor cu frecventa 1(list)
'''

def Cuv1(sir):
  text = sir.split()
  frecv = {}
  for cuv in text:
    if cuv in frecv:
      frecv[cuv] += 1
    else:
      frecv[cuv] = 1
  rez = []
  for key, value in frecv.items():
    if value == 1:
      rez.append(key)
  return rez

#folosesc un dictionar
#O(n*log(n))
'''
input: sir de caractere(string)
output: lista cuvintelor cu frecventa 1(lista de string-uri)
'''

def Cuv2(sir):
  text = sir.split()
  lista =  sorted(text)
  rez = []
  aparitii = defaultdict(list)
  for cuv in lista:
    aparitii[cuv].append(cuv)
  for cuv,list_ap in aparitii.items():
    if len(list_ap) == 1:
      rez.append(list_ap[0])
  return rez

#testare

def Test4(cerinta):
  print("Test Start")
  assert(len(cerinta('ana are ana are mere rosii ana'))==2)
  assert(len(cerinta('ana are ana are mere mere rosii rosii ana'))==0)
  assert(len(cerinta('ana are mere'))==3)
  print("Succes!")

#O(n)
#n numar cuvinte
# Impartim textul in cuvinte
# Cream un dictionar pentru a stoca frecventa cuvintelor
# Numaram frecventa fiecarui cuvant
# Convertim la litere mici
# Gasim cuvintele care apar exact o singura data

def Copilot(text):
  cuvinte = text.split()
  frecventa_cuvinte = {}
  for cuvant in cuvinte:
      cuvant = cuvant.lower()
      if cuvant in frecventa_cuvinte:
          frecventa_cuvinte[cuvant] += 1
      else:
          frecventa_cuvinte[cuvant] = 1
  cuvinte_unice = [cuvant for cuvant, frecventa in frecventa_cuvinte.items() if frecventa == 1]
  return cuvinte_unice

Test4(Cuv1)
Test4(Cuv2)
Test4(Copilot)

sir = "ana are ana are mere rosii ana"
text = Cuv1(sir)
text1 = Cuv2(sir)

print("Cuvi care apar o singura data:")
for i in text:
  print(i)
print("Cuv care apar o singura data", text1)

#Copilot

print(f"Copilot: Cuvintele unice din text sunt: {', '.join(Copilot(sir))}")
```

```python
#5)Pentru un șir cu n elemente care conține valori din mulțimea {1, 2, ..., n - 1} astfel
#încât o singură valoare se repetă de două ori, să se identifice acea valoare care se repetă.
#De ex. în șirul [1,2,3,4,2] valoarea 2 apare de două ori.

'''
creez un vector de frecventa si cand acesta creste la 2 ii afisez cheia
O(n)

input: sir de caractere(string)
output: valoarea care se repeta(int)
'''

def Rep1(l):
  frecv = {}
  nr = 0
  for i in l:
    if i != " ":
      if i in frecv :
        frecv[i]+=1
      else:
        frecv[i]=1
      if frecv[i] == 2:
        nr = i
        break
  return nr

'''
Parcurgere lista de numere, daca nu exista in set il adaug, daca exista in set afisez
acel numar
O(n)

input: sir de caractere(string)
output: valoarea care se repeta(int)
'''

def Rep2(l):
  nr = set()
  for i in l:
    if i != " ":
      if i in nr:
        return i
      else:
        nr.add(i)
  return 0

#testare

def Test5(cerinta):
  print("Test Start")
  assert cerinta([1,2,3,4,2]) == 2
  print("Succes")

#O(n)
#Calculam suma ele din sir
#Calculam suma primelor n-1 nr nat
#Diferenta dintre sumele e val care se repeta

def Copilot(nums):
  n = len(nums)
  sumElem = sum(nums)
  sumNat = (n-1)*n//2
  duplicate = sumElem - sumNat
  return duplicate

Test5(Rep1)
Test5(Rep2)
Test5(Copilot)

l = "1 2 3 2"
print("Nr care apare de 2 ori este: ",Rep1(l))
print("Nr care apare de 2 ori este: ",Rep2(l))

#Copilot

arr = [1,2,3,4,2]
duplicate = Copilot(arr)
print("Copilot: Valoarea care se repetă în șirul", arr, "este:", duplicate)
```

```python
from collections import defaultdict

#6) Pentru un șir cu n numere întregi care conține și duplicate, să se determine elementul
#majoritar (care apare de mai mult de n / 2 ori). De ex. 2 este elementul majoritar în șirul
# [2,8,7,2,2,5,2,3,1,2,2].

#folosim vector de frecventa
#O(n)
'''
input: sir de caractere(string)
output: valoarea care se repeta(int)

'''
def Max1(l):
  frecv = {}
  nr =  0
  for i in l:
    if i in frecv:
      frecv[i] += 1
    else:
      frecv[i] = 1
  lung = int(len(l)/2)
  rez = set()
  for key,value in frecv.items():
    if value >= lung:
      rez.add(key)
  return rez

#folosim un dictionar
#O(n*log(n))
'''
input: sir de caractere(string)
output: valoarea care se repeta(int)

'''

def Max2(l):
  lista = sorted(l)
  lung = int(len(l)/2)
  rez = []
  aparitii = defaultdict(list)
  for nr in lista:
    aparitii[nr].append(nr)
  for nr,apare in aparitii.items():
    if len(apare) >= lung:
      rez.append(nr)
  return rez

#testare

def Test6(cerinta):
  print("Testare start")
  if cerinta == "Copilot":
    assert(cerinta([2,8,7,2,2,5,2,3,1,2,2]) == 2)
  else:
    assert(cerinta([1,1,2,2,3]) == {1,2})
    assert(cerinta([2,8,7,2,2,5,2,3,1,2,2]) == {2})
  print("Succes")

def Test6_2(cerinta):
  print("Testare start")
  if cerinta == "Copilot":
    assert(cerinta([2,8,7,2,2,5,2,3,1,2,2]) == 2)
  else:
    assert(cerinta([1,1,2,2,3]) == [1,2])
    assert(cerinta([2,8,7,2,2,5,2,3,1,2,2]) == [2])
  print("Succes")

#O(n)
# Verificăm dacă candidatul este de fapt element majoritar
def Copilot(nums):
    candidate = None
    count = 0
    for num in nums:
        if count == 0:
            candidate = num
            count = 1
        elif candidate == num:
            count += 1
        else:
            count -= 1
    count = 0
    for num in nums:
        if num == candidate:
            count += 1
    if count > len(nums) / 2:
        return candidate
    else:
        return None

Test6(Max1)
Test6_2(Max2)
#Test6_2(Copilot)

l = [1,1,2,2,3]
print("Rasp pt [2,8,7,2,2,5,2,3,1,2,2]: ", Max1(l))
print("Rasp pt [2,8,7,2,2,5,2,3,1,2,2]: ", Max2(l))

arr = [2,8,7,2,2,5,2,3,1,2,2]
maj_elem = Copilot(arr)
if maj_elem is not None:
   print("Copilot: Elementul majoritar în șirul", arr, "este:", maj_elem)
else:
    print("Copilot: Nu există element majoritar în șirul", arr)
```
