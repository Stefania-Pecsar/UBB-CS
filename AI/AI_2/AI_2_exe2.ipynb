```python
#b)daca imaginile nu au aceeasi dimensiune, sa se redimensioneze toate la 128 x 128 pixeli
#si sa se vizualizeze imaginile intr-un cadru tabelar.
from PIL import Image

#incarcare poze
img1 = Image.open('Altman.webp')
img2 = Image.open('BERT.png')
img3 = Image.open('chatGPT.png')
img4 = Image.open('diffusionModel.jpg')
img5 = Image.open('Karpaty.jpg')
img6 = Image.open('LeCun.jpg')
img7 = Image.open('Leskovec.jpg')
img8 = Image.open('Ng.webp')
img9 = Image.open('Norvig.jpg')
img10 = Image.open('Russell.jpg')
img11 = Image.open('Turing.webp')
img12 = Image.open('YOLO.jpg')

imagini=[img1,img2,img3,img4,img5,img6,img7,img8,img9,img10,img11,img12]
#obtinem dimensiunea primei imagini
width, height = imagini[0].size

#vf daca imag au aceiasi dim
aceeasi_dimensiune = True
for i in imagini:
    if(i.size != (width, height)):
        aceeasi_dimensiune = False
        break

if not aceeasi_dimensiune:
    imgNou1 = img1.resize((128,128))
    imgNou2 = img2.resize((128,128))
    imgNou3 = img3.resize((128,128))
    imgNou4 = img4.resize((128,128))
    imgNou5 = img5.resize((128,128))
    imgNou6 = img6.resize((128,128))
    imgNou7 = img7.resize((128,128))
    imgNou8 = img8.resize((128,128))
    imgNou9 = img9.resize((128,128))
    imgNou10 = img10.resize((128,128))
    imgNou11 = img11.resize((128,128))
    imgNou12 = img12.resize((128,128))

    fig, axs = plt.subplots(3, 4, figsize=(12, 9))
    axs[0, 0].imshow(imgNou1)
    axs[0, 1].imshow(imgNou2)
    axs[0, 2].imshow(imgNou3)
    axs[0, 3].imshow(imgNou4)
    axs[1, 0].imshow(imgNou5)
    axs[1, 1].imshow(imgNou6)
    axs[1, 2].imshow(imgNou7)
    axs[1, 3].imshow(imgNou8)
    axs[2, 0].imshow(imgNou9)
    axs[2, 1].imshow(imgNou10)
    axs[2, 2].imshow(imgNou11)
    axs[2, 3].imshow(imgNou12)
else:
    print("Imaginile nu au suferit nicio redimensionare")
plt.show()
```

```python
#c)sa se transforme imaginile in format gray-levels si sa se vizualizeze

img1 = Image.open('Altman.webp')
img2 = Image.open('BERT.png')
img3 = Image.open('chatGPT.png')
img4 = Image.open('diffusionModel.jpg')
img5 = Image.open('Karpaty.jpg')
img6 = Image.open('LeCun.jpg')
img7 = Image.open('Leskovec.jpg')
img8 = Image.open('Ng.webp')
img9 = Image.open('Norvig.jpg')
img10 = Image.open('Russell.jpg')
img11 = Image.open('Turing.webp')
img12 = Image.open('YOLO.jpg')

imgGri1 = img1.convert('L')
imgGri2 = img2.convert('L')
imgGri3 = img3.convert('L')
imgGri4 = img4.convert('L')
imgGri5 = img5.convert('L')
imgGri6 = img6.convert('L')
imgGri7 = img7.convert('L')
imgGri8 = img8.convert('L')
imgGri9 = img9.convert('L')
imgGri10 = img10.convert('L')
imgGri11 = img11.convert('L')
imgGri12 = img12.convert('L')

# Adugarea imaginilor ntr-o list
imagini_grayscale = [imgGri1, imgGri2, imgGri3, imgGri4, imgGri5, imgGri6, imgGri7, imgGri8, imgGri9, imgGri10, imgGri11, imgGri12]

#creare grid pt viziualizare
fig, axs = plt.subplots(3, 4, figsize=(15, 9))
for i, ax in enumerate(axs.flat):
    ax.imshow(imagini_grayscale[i], cmap='gray')  # Afieaz imaginea n grayscale
    ax.axis('off')

plt.show()
```

```python
#d)sa se blureze o imagine si sa se afiseze in format "before-after"
from PIL import Image, ImageFilter
#pe imag redimensionata la 128 x 128 aplicam blur
img_blurred = imgNou4.filter(ImageFilter.BLUR)
#grafic pt vizualizare
fig, axs = plt.subplots(1, 2, figsize=(15, 9))
axs[0].imshow(imgNou4)
axs[1].imshow(img_blurred)
plt.show()
```
